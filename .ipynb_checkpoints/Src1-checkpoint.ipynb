```python
#Import the necessary libraries 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 

# Load the image 
image = cv2.imread('C://Users//royad//OneDrive//Documents//6thSem//CGMiniProject//ImagetoRestoreWithF.jpeg') 

#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

# Adjust the brightness and contrast 
# Adjusts the brightness by adding each pixel value 
brightness =3.4
# Adjusts the contrast by scaling the pixel values  
contrast = 1.2
image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness) 
 
#Plot the contrast image 
plt.subplot(1, 2, 2) 
plt.title("Brightness & contrast") 
plt.imshow(image2) 
plt.show()

# Create the sharpening kernel
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) 
  
# Sharpen the image 
sharpened_image = cv2.filter2D(image2, -1, kernel) 

#Save the image 
cv2.imwrite('sharpened_image.jpg', sharpened_image)

#Plot the sharpened image 
plt.subplot(1, 2, 2) 
plt.title("Sharpening") 
plt.imshow(sharpened_image) 
plt.show()

# Create a color map (e.g., apply a colormap for pseudo-coloring)
colored_image = cv2.applyColorMap(sharpened_image, cv2.COLORMAP_OCEAN)

#Plot the enhanced image 
plt.subplot(1, 2, 2) 
plt.title("enhanced coloured") 
plt.imshow(colored_image) 
plt.show()
```
