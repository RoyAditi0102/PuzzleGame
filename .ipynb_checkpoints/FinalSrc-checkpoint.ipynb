```python
import cv2
import numpy as np
import random
import pygame
import time

# Pygame Initialization
pygame.init()
pygame.mixer.init()

# Define Colors
BLACK = (0, 0, 0)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
WHITE = (255, 255, 255)
ORANGE = (255, 165, 0)

# Define Font Size
FONT_SIZE = 30

# Define Screen Dimensions
WIDTH = 800
HEIGHT = 600

# Load Background Music
pygame.mixer.music.load('background.mp3')  # Replace with your music file

# Create the Pygame Screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))

# List of image paths
image_paths = [
    'image1.jpeg.jpg',  # Replace with actual image paths
    'image2.jpeg.jpg',
    'image3.jpeg.jpg',
    'image4.jpeg.jpg',
    'image5.jpeg.jpg',
    'image6.jpeg.jpg',
    'image7.jpeg.jpg',
    'image8.jpeg.jpg',
    'image9.jpeg.jpg',
    'image10.jpeg.jpg'
]

# Function to Draw Centered Text
def draw_centered_text(surface, text, size, color, y):
    font_size = int(size)  # Ensure font size is an integer
    font = pygame.font.SysFont(None, font_size)
    text_surface = font.render(text, True, color)
    text_rect = text_surface.get_rect(center=(WIDTH // 2, y))
    surface.blit(text_surface, text_rect)

# Function to Draw Text
def draw_text(surface, text, size, color, position):
    font_size = int(size)  # Ensure font size is an integer
    font = pygame.font.SysFont(None, font_size)
    text_surface = font.render(text, True, color)
    surface.blit(text_surface, position)

# Function to Show Introduction Screen
def show_intro_screen():
    pygame.mixer.music.play(-1)  # Loop background music
    screen.fill(BLACK)
    draw_centered_text(screen, 'SAPTHAGIRI COLLEGE OF ENGINEERING', int(FONT_SIZE * 1.5), RED, HEIGHT // 10)
    draw_centered_text(screen, 'DEPARTMENT OF COMPUTER SCIENCE AND ENGINEERING', FONT_SIZE, GREEN, HEIGHT // 8 + 25)
    draw_centered_text(screen, 'COMPUTER GRAPHICS AND VISUALIZATION', FONT_SIZE, GREEN, HEIGHT // 6 + 40)
    draw_centered_text(screen, 'MINI PROJECT ON', FONT_SIZE, BLUE, HEIGHT // 5 + 60)
    draw_centered_text(screen, 'SLIDING PUZZLE', int(FONT_SIZE * 1.5), BLUE, HEIGHT // 4 + 80)

    draw_text(screen, 'TEAM MEMBERS', FONT_SIZE, WHITE, (20, HEIGHT // 2.2))
    draw_text(screen, 'AYUSHI GUPTA :1SG21CS012', FONT_SIZE, ORANGE, (20, HEIGHT // 2 + 20))
    draw_text(screen, 'ADITI ROY : 1SG21CS003', FONT_SIZE, ORANGE, (20, HEIGHT // 2 + 60))
    draw_text(screen, 'ANINDITA GOSWAMI : 1SG21CS006', FONT_SIZE, ORANGE, (20, HEIGHT // 2 + 100))

    draw_text(screen, 'PROJECT GUIDED BY', FONT_SIZE, WHITE, (WIDTH // 2 + 100, HEIGHT // 2.2))
    draw_text(screen, 'Prof. SHEELARANI C M', FONT_SIZE, ORANGE, (WIDTH // 2 + 100, HEIGHT // 2 + 20))
    draw_text(screen, 'Assistant Professor', FONT_SIZE, ORANGE, (WIDTH // 2 + 100, HEIGHT // 2 + 60))
    draw_text(screen, 'Dept. of CS&E', FONT_SIZE, ORANGE, (WIDTH // 2 + 100, HEIGHT // 2 + 100))

    draw_centered_text(screen, 'Press ENTER to Start', FONT_SIZE, RED, HEIGHT - 50)

    pygame.display.flip()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                raise SystemExit
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RETURN:
                    pygame.mixer.music.stop()  # Stop background music
                    return
                if event.key == pygame.K_ESCAPE:
                    pygame.mixer.music.stop()  # Stop background music
                    pygame.quit()
                    raise SystemExit

# Function to Show Instruction Screen
def show_instruction_screen():
    screen.fill(BLACK)
    draw_centered_text(screen, 'INSTRUCTIONS', int(FONT_SIZE * 1.5), RED, HEIGHT // 10)
    draw_text(screen, 'Use the following keys to play the game:', FONT_SIZE, WHITE, (20, HEIGHT // 4))
    draw_text(screen, 'W - Move up', FONT_SIZE, GREEN, (20, HEIGHT // 4 + 40))
    draw_text(screen, 'A - Move left', FONT_SIZE, GREEN, (20, HEIGHT // 4 + 80))
    draw_text(screen, 'S - Move down', FONT_SIZE, GREEN, (20, HEIGHT // 4 + 120))
    draw_text(screen, 'D - Move right', FONT_SIZE, GREEN, (20, HEIGHT // 4 + 160))
    draw_text(screen, 'ESC - Exit the game', FONT_SIZE, GREEN, (20, HEIGHT // 4 + 200))
    draw_text(screen, 'R - Restart the game', FONT_SIZE, GREEN, (20, HEIGHT // 4 + 240))
    draw_centered_text(screen, 'Press ENTER to Start the Game', FONT_SIZE, RED, HEIGHT - 50)

    pygame.display.flip()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                raise SystemExit
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RETURN:
                    return
                if event.key == pygame.K_ESCAPE:
                    pygame.quit()
                    raise SystemExit

# Function to Show Congratulatory Screen
def show_congrats_screen():
    screen.fill(BLACK)
    draw_centered_text(screen, 'Congratulations!', int(FONT_SIZE * 1.5), RED, HEIGHT // 2)
    draw_centered_text(screen, 'You have successfully solved the puzzle!', FONT_SIZE, GREEN, HEIGHT // 2 + 50)
    draw_centered_text(screen, 'Press ESC to Exit', FONT_SIZE, RED, HEIGHT - 50)

    pygame.display.flip()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                raise SystemExit
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    pygame.quit()
                    raise SystemExit

# Function to Show Time Up Screen
def show_time_up_screen():
    screen.fill(BLACK)
    draw_centered_text(screen, 'Oops, time up!', int(FONT_SIZE * 1.5), RED, HEIGHT // 2)
    draw_centered_text(screen, 'Better luck next time!', FONT_SIZE, GREEN, HEIGHT // 2 + 50)
    draw_centered_text(screen, 'Press ESC to Exit', FONT_SIZE, RED, HEIGHT - 50)

    pygame.display.flip()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                raise SystemExit
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    pygame.quit()
                    raise SystemExit

# OpenCV Sliding Puzzle Game
def divide_image(image, rows, cols):
    tiles = []
    h, w, _ = image.shape
    tile_height = h // rows
    tile_width = w // cols
    for r in range(rows):
        row_tiles = []
        for c in range(cols):
            tile = image[r*tile_height:(r+1)*tile_height, c*tile_width:(c+1)*tile_width]
            row_tiles.append(tile)
        tiles.append(row_tiles)
    return tiles

def shuffle_tiles(tiles):
    flat_tiles = sum(tiles, [])
    random.shuffle(flat_tiles)
    shuffled_tiles = [flat_tiles[i:i+len(tiles[0])] for i in range(0, len(flat_tiles), len(tiles[0]))]
    return shuffled_tiles

def draw_tiles(tiles, tile_height, tile_width):
    rows = len(tiles)
    cols = len(tiles[0])
    canvas = np.zeros((tile_height*rows, tile_width*cols, 3), dtype=np.uint8)
    for r in range(rows):
        for c in range(cols):
            if tiles[r][c] is not None:
                canvas[r*tile_height:(r+1)*tile_height, c*tile_width:(c+1)*tile_width] = tiles[r][c]
    return canvas

def find_blank_tile(tiles):
    for r in range(len(tiles)):
        for c in range(len(tiles[r])):
            if tiles[r][c] is None:
                return r, c
    return None

def move_tile(tiles, r, c, blank_r, blank_c):
    tiles[blank_r][blank_c], tiles[r][c] = tiles[r][c], tiles[blank_r][blank_c]

def is_solved(tiles, original_tiles):
    for r in range(len(tiles)):
        for c in range(len(tiles[r])):
            if (tiles[r][c] is not None) and not np.array_equal(tiles[r][c], original_tiles[r][c]):
                return False
    return True

def reset_puzzle():
    global tiles
    tiles = shuffle_tiles(original_tiles)
    tiles[-1][-1] = None  # Ensure the last tile is blank
    return tiles

# Initialize
image_path = random.choice(image_paths)  # Choose a random image path
image = cv2.imread(image_path)
if image is None:
    print("Error: Unable to load image. Check the file path and try again.")
else:
    show_intro_screen()  # Show the introduction screen before starting the puzzle
    show_instruction_screen()  # Show the instruction screen before starting the puzzle

    # Scale image to fit within the screen dimensions
    aspect_ratio = image.shape[1] / image.shape[0]
    if WIDTH / aspect_ratio <= HEIGHT:
        desired_width = WIDTH
        desired_height = int(WIDTH / aspect_ratio)
    else:
        desired_height = HEIGHT
        desired_width = int(HEIGHT * aspect_ratio)

    resized_image = cv2.resize(image, (desired_width, desired_height))

    rows, cols = 3, 3  # Define the grid size for 3x3
    original_tiles = divide_image(resized_image, rows, cols)
    tiles = shuffle_tiles(original_tiles)
    
    # Set the last tile as blank (None)
    tiles[-1][-1] = None
    blank_r, blank_c = rows - 1, cols - 1

    tile_height, tile_width, _ = original_tiles[0][0].shape

    # Start the timer
    start_time = time.time()

    while True:
        elapsed_time = time.time() - start_time
        if elapsed_time > 60:  # 1 minute timer
            cv2.destroyAllWindows()  # Close the OpenCV window
            show_time_up_screen()  # Show the time-up screen using Pygame
            break
        
        canvas = draw_tiles(tiles, tile_height, tile_width)

        # Display the timer on the top-right corner
        timer_text = f"Time: {int(elapsed_time)}s"
        cv2.putText(canvas, timer_text, (canvas.shape[1] - 150, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        if is_solved(tiles, original_tiles):
            cv2.destroyAllWindows()  # Close the OpenCV window
            show_congrats_screen()  # Show the congratulatory screen using Pygame
            break

        cv2.imshow('Sliding Puzzle', canvas)

        key = cv2.waitKey(100) & 0xFF
        if key == 27:  # Esc key
            cv2.destroyAllWindows()
            break
        elif key == ord('r'):
            tiles = reset_puzzle()
            blank_r, blank_c = rows - 1, cols - 1
            start_time = time.time()  # Reset the timer
        elif key in [ord('w'), ord('a'), ord('s'), ord('d')]:
            blank_r, blank_c = find_blank_tile(tiles)
            if key == ord('w') and blank_r < rows - 1:
                move_tile(tiles, blank_r + 1, blank_c, blank_r, blank_c)
            elif key == ord('a') and blank_c < cols - 1:
                move_tile(tiles, blank_r, blank_c + 1, blank_r, blank_c)
            elif key == ord('s') and blank_r > 0:
                move_tile(tiles, blank_r - 1, blank_c, blank_r, blank_c)
            elif key == ord('d') and blank_c > 0:
                move_tile(tiles, blank_r, blank_c - 1, blank_r, blank_c)

    pygame.quit()  # Ensure Pygame quits
```
